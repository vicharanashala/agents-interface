```python
import pandas as pd
import os

# Paths
file_path = '/home/aic_u2/aic_u2/Shubhankar/answer_generation/notebook/Agri Data across Regions Sheet Google (1).xlsx'
output_dir = '/home/aic_u2/aic_u2/Shubhankar/answer_generation/notebook/csvs_data'

# Create output folder if not exists
os.makedirs(output_dir, exist_ok=True)

# Read all sheets
all_sheets = pd.read_excel(file_path, sheet_name=None)

# Dictionary to group sheets by header tuple
header_groups = {}

# Group sheets that have the same headers
for sheet_name, df in all_sheets.items():
    headers = tuple(df.columns)
    header_groups.setdefault(headers, []).append((sheet_name, df))

# Merge sheets within each header group and save as CSV
for i, (headers, sheets) in enumerate(header_groups.items(), start=1):
    merged_df = pd.DataFrame()
    for sheet_name, df in sheets:
        df = df.copy()
        df["sheet_name"] = sheet_name
        merged_df = pd.concat([merged_df, df], ignore_index=True)
    
    # Create a descriptive filename (or just group number)
    file_name = f"group_{i}.csv"
    output_path = os.path.join(output_dir, file_name)
    
    merged_df.to_csv(output_path, index=False)
    print(f"Saved {output_path} with sheets: {[s[0] for s in sheets]}")
```

```python
import pandas as pd

# Assuming summary_df already exists and looks like:
# | sheet_name | headers |
# |-------------|----------|

# Convert header lists to tuples so they can be compared
summary_df['header_signature'] = summary_df['headers'].apply(lambda x: tuple(x))

# Group by identical header sets
grouped = summary_df.groupby('header_signature')['sheet_name'].apply(list).reset_index()

# Keep only those groups where more than one sheet shares the same headers
similar_headers = grouped[grouped['sheet_name'].apply(lambda x: len(x) > 1)]

print("Sheets with identical headers:\n")
for _, row in similar_headers.iterrows():
    print(f"{row['sheet_name']} share the same headers:")
    print(list(row['header_signature']))
    print("-" * 80)
```

```python
import pandas as pd
import google.generativeai as genai
import time
from tqdm import tqdm
import os

# Configure Gemini API
# Replace 'YOUR_API_KEY' with your actual Gemini API key
genai.configure(api_key='AIzaSyDDCD9cNSrjwup1XpYbiBL9Dqp2OJVUZXc')

# Initialize the model
model = genai.GenerativeModel('gemini-2.5-flash')

```

```python

# Read the q&a.csv file with all context columns
df = pd.read_csv('/home/aic_u2/aic_u2/Shubhankar/answer_generation/notebook/csvs_data/q&a.csv')

# Take only first 100 questions
df_100 = df.head(100).copy()

# Output file path
output_file = '/home/aic_u2/aic_u2/Shubhankar/answer_generation/notebook/csvs_data/q&a100.csv'

# Start fresh - create file with headers
first_row = True

# Iterate through each question
for idx, row in tqdm(df_100.iterrows(), total=len(df_100), desc="Processing questions"):
    question_id = row['id']
    question = row['QUESTION [by Agri Team]']
    
    # Build context from additional columns
    context_parts = []
    if pd.notna(row.get('Category')):
        context_parts.append(f"Category: {row['Category']}")
    if pd.notna(row.get('Month')):
        context_parts.append(f"Month: {row['Month']}")
    if pd.notna(row.get('Crop')):
        context_parts.append(f"Crop: {row['Crop']}")
    if pd.notna(row.get('State')):
        context_parts.append(f"State: {row['State']}")
    if pd.notna(row.get('QueryType')):
        context_parts.append(f"Query Type: {row['QueryType']}")
    if pd.notna(row.get('Season')):
        context_parts.append(f"Season: {row['Season']}")
    
    context = "\n".join(context_parts)
    
    # Create prompt with context
    prompt = f"""You are an agricultural expert. Please answer the following question based on the context provided.

Context:
{context}

Question: {question}

Provide a detailed, accurate, and practical answer for farmers."""
    
    try:
        # Generate answer using Gemini
        response = model.generate_content(prompt)
        answer = response.text
        
        # Store the result
        result_row = {
            'id': question_id,
            'question': question,
            'answer': answer
        }
        
        # Append to CSV immediately
        result_df = pd.DataFrame([result_row])
        if first_row:
            result_df.to_csv(output_file, mode='w', header=True, index=False)
            first_row = False
        else:
            result_df.to_csv(output_file, mode='a', header=False, index=False)
        
        print(f"✓ Processed & Saved {idx+1}/100 - ID: {question_id}")
        
        # Add a small delay to avoid rate limiting
        time.sleep(1)
        
    except Exception as e:
        print(f"✗ Error processing ID {question_id}: {str(e)}")
        
        # Save error to CSV as well
        result_row = {
            'id': question_id,
            'question': question,
            'answer': f"ERROR: {str(e)}"
        }
        result_df = pd.DataFrame([result_row])
        if first_row:
            result_df.to_csv(output_file, mode='w', header=True, index=False)
            first_row = False
        else:
            result_df.to_csv(output_file, mode='a', header=False, index=False)
        
        time.sleep(2)  # Longer delay after error

# Read final CSV and display summary
final_df = pd.read_csv(output_file)
print(f"\n{'='*50}")
print(f"✓ Complete! Total saved: {len(final_df)} question-answer pairs")
print(f"{'='*50}")
print(f"\nFirst few rows:")
print(final_df.head())
```
